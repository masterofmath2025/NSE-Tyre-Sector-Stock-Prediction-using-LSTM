```python
def insertVariblesIntoTable(Date, Open, Low, High, Close, Adj_close,Volume):
    try:
        connection = connector.connect(host='localhost',
                                             database='stocks',
                                             user='root',
                                             password='1998')
        cursor = connection.cursor()
        mySql_insert_query = '''INSERT INTO apollo (Date, Open, Low, High, Close, Adj_Close,Volume) 
                                VALUES (%s, %s, %s, %s, %s, %s,%s)'''

        recordTuple = (Date, Open, Low, High, Close, Adj_close,Volume)
        cursor.execute(mySql_insert_query, recordTuple)
        connection.commit()
        print("Record inserted successfully into Laptop table")

    except connector.Error as error:
        print("Failed to insert into MySQL table {}".format(error))

    finally:
        if (connection.is_connected()):
            cursor.close()
            connection.close()
            print("MySQL connection is closed")

```

```python
import yfinance as yf
import time
```

```python
minutes = 1
threshold = len(yf.download("APOLLOTYRE.NS",period="1d",interval="1m"))
print("threshold:",threshold)
value = %sql SELECT COUNT('Date') FROM apollo
while True:
    if value[0][0] < threshold:
        ################################################################################
        raydaily = yf.download("APOLLOTYRE.NS",period="1d",interval="1m")

        insertflag = len(raydaily)
        print("insertflag:",insertflag)
                ############################################################################
        ############################################################################
        raydaily["Volume"] = raydaily["Volume"].astype(float)
        for rows in range(0,threshold):
            insertVariblesIntoTable(raydaily.index[rows],raydaily["Open"][rows],
                                    raydaily["Low"][rows],raydaily["High"][rows],
                                    raydaily["Close"][rows],raydaily["Adj Close"][rows],raydaily["Volume"][rows])
            
        #time.sleep(60) # Delay for every 60 secs (60 seconds).
        value = %sql SELECT COUNT('Date') FROM apollo

        #print("Time passed {} minutes".format(minutes))
        #minutes += 1
        
        
    else:
        #################################################################################
        raydaily = yf.download("APOLLOTYRE.NS",period="1d",interval="1m")

        insertflag = len(raydaily)
        print("insertflag:",insertflag)
        
        ##################################################################################
        ##################################################################################
        raydaily["Volume"] = raydaily["Volume"].astype(float)
        if insertflag > value[0][0]:
            extra = insertflag - value[0][0]
            for rows in range(-(extra),0):
                insertVariblesIntoTable(raydaily.index[rows],raydaily["Open"][rows],
                                        raydaily["Low"][rows],raydaily["High"][rows],
                                        raydaily["Close"][rows],raydaily["Adj Close"][rows],raydaily["Volume"][rows])


        time.sleep(60) # Delay for every 60 secs (60 seconds).
        value = %sql SELECT COUNT('Date') FROM apollo

        print("Time passed {} minutes".format(minutes))
        minutes += 1

```
